```python
#Change the column name for better understanding
new_cols_dict ={
    "timestamp" : "time",
    "cnt" : "count",
    "t1" : "temp_real_C",
    "t2" : "temp_feels_like_C",
    "hum": "humidity_percent",
    "wind_speed": "wind_speed_kph",
    "weather_code": "weather",
    "is_holiday" : "is_holiday",
    "is_weekend" : "is_weekend",
    "season" : "season"

}

# Reanaming the columns to the specified column names
bikes.rename(new_cols_dict, axis=1, inplace=True)
bikes
```

```python
# Changing the humidity values to percentage (i.e. a value between 0 and 1)
bikes.humidity_percent = bikes.humidity_percent / 100    
bikes
```

```python
# Mapping integers to actual season names
season_dict = { 
    '0.0': 'spring',
    '1.0': 'summer',
    '2.0': 'autumn',
    '3.0': 'winter'
}

# Mapping integers to actual weather conditions
weather_dict = {
    '1.0': 'Clear',
    '2.0': 'Scattered clouds',
    '3.0': 'Broken clouds',
    '4.0': 'Cloudy',
    '7.0': 'Rain',
    '10.0': 'Rain with thunderstorm',
    '26.0': 'Snowfall'
}

# Changing the seasons and weather columns to string and mapping values
bikes['season'] = bikes['season'].astype('str').map(season_dict)

bikes['weather'] = bikes['weather'].astype('str').map(weather_dict)

bikes
```
